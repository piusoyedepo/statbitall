# Hypothesis testing from scratch: the logic before the formula
[Read the full post on Statbitall](https://statbitall.com/blog/b0-hypothesis-testing-logic)

```python
import numpy as np
from scipy.stats import t as t_dist

np.random.seed(42)
n = 40
sample = np.random.normal(loc=1045, scale=120, size=n)

mu_0 = 1000
alpha = 0.05
```

```python
x_bar = np.mean(sample)
s = np.std(sample, ddof=1)
se = s / np.sqrt(n)
t_stat = (x_bar - mu_0) / se

df = n - 1
p_value = 2 * t_dist.sf(np.abs(t_stat), df=df)

print(f"Sample mean:    {x_bar:.2f}")
print(f"Test statistic: t = {t_stat:.4f}")
print(f"p-value:        {p_value:.4f}")
print(f"Decision:       {'Reject H0' if p_value < alpha else 'Fail to reject H0'}")
```
